# Database Connection

# Function to Load SQL Files

```python
def load_sql_query(file_path):

    with open(file_path, "r", encoding="utf-8") as file:
        sql = file.read()

    return sql
```

# Load Main EDA Dataset

```python
sql = load_sql_query(
    "../database/EDA/01_sales_dataset.sql"
)

sales_df = pd.read_sql(sql, engine)

sales_df.head()
```

# Revenue Overview

```python
sales_df["payment_value"].agg(
    ["count", "sum", "mean", "min", "max"]
)
```

# Top Categories

```python
(
    sales_df
    .groupby("category")["payment_value"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

# Top States

```python
(
    sales_df
    .groupby("customer_state")["payment_value"]
    .sum()
    .sort_values(ascending=False)
)
```
